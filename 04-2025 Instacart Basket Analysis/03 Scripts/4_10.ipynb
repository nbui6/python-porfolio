```python
# 5. Create a profiling variable based on age, income, certain goods in the “department_id” column, and number of dependents

customer_df = all_data_merge.groupby('user_id').agg({
    'age_group': 'first',
    'income_group': 'first',
    'n_dependants': 'max',
    'department_id': lambda x: list(set(x)),  # departments they've bought from
    'order_hour_of_day': lambda x: list(x),
    'orders_day_of_week': lambda x: list(x),
    'region': 'first',
    'gender': 'first'
}).reset_index()
```

```python
# How many unique departments a customer buys from
def count_unique_departments(dept_list):
    return len(set(dept_list))
customer_df['num_departments'] = customer_df['department_id'].apply(count_unique_departments)
def department_type(n):
    if n >= 10:
        return 'Diverse'
    elif n >= 5:
        return 'Medium'
    else:
        return 'Focused'
customer_df['department_type'] = customer_df['num_departments'].apply(department_type)

# Morning person?
def morning_ratio(hour_list):
    if not hour_list:  # if the list is empty
        return 0
    morning_count = 0
    for hour in hour_list:
        if 6 <= hour <= 11:
            morning_count += 1
    return morning_count / len(hour_list)  # percent of orders in morning
customer_df['morning_shopper'] = customer_df['order_hour_of_day'].apply(morning_ratio)
customer_df['is_morning_shopper'] = customer_df['morning_shopper'] >= 0.5

# Weekend shopper?
def weekend_ratio(day_list):
    if not day_list:
        return 0
    weekend_count = 0
    for day in day_list:
        if day in [5, 6]:
            weekend_count += 1
    return weekend_count / len(day_list)
customer_df['weekend_shopper'] = customer_df['orders_day_of_week'].apply(weekend_ratio)
customer_df['is_weekend_shopper'] = customer_df['weekend_shopper'] >= 0.5

# With dependants? 
customer_df['has_dependants'] = customer_df['n_dependants'] > 0
```

```python
# Combine categorical features
customer_df_grouped = customer_df.groupby(['age_group', 'income_group', 'has_dependants', 'department_type', 'is_morning_shopper', 'is_weekend_shopper']).size().reset_index(name='count')
# Sort to see largest groups
customer_df_grouped = customer_df_grouped.sort_values('count', ascending=False)
customer_df_grouped
```

```python
# generate dictionary of 10 top groups
top_profiles = {
    # 60+ AGE GROUP
    ('60+', 'High', True, 'Diverse', False, False): 'Retired Family Planner',
    ('60+', 'High', True, 'Medium', False, False): 'Simplified Senior Household',
    ('60+', 'High', False, 'Diverse', False, False): 'Independent Senior Shopper',
    ('60+', 'Mid', True, 'Diverse', False, False): 'Traditional Mid-Income Retiree',

    # 41–60 AGE GROUP
    ('41–60', 'High', True, 'Diverse', False, False): 'Affluent Traditional Household',
    ('41–60', 'High', True, 'Medium', False, False): 'Medium-Range Affluent Family',
    ('41–60', 'High', False, 'Diverse', False, False): 'Affluent Independent Planner',
    ('41–60', 'Mid', True, 'Diverse', False, False): 'Balanced Family Planner',

    # 26–40 AGE GROUP
    ('26–40', 'Mid', True, 'Diverse', False, False): 'Budget-Conscious Young Family',

    # 18–25 AGE GROUP
    ('18–25', 'Mid', True, 'Diverse', False, False): 'Young Parent Essentials Buyer',
}
```

```python
def get_profile(row):
    key = (
        row['age_group'],
        row['income_group'],
        row['has_dependants'],
        row['department_type'],
        row['is_morning_shopper'],
        row['is_weekend_shopper']
    )
    return top_profiles.get(key, 'Other')

customer_df['profile'] = customer_df.apply(get_profile, axis=1)
```
