```python

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'

```

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean']})
```

```python
# spending flag for each user based on the average price across all their orders
user_avg = ords_prods_merge.groupby('user_id')['prices'].mean().reset_index()
```

```python
user_avg.loc[user_avg['prices'] < 10, 'high_spender'] = 'Low Spender'
user_avg.loc[user_avg['prices'] >= 10, 'high_spender'] = 'High Spender'
ords_prods_merge = ords_prods_merge.merge(user_avg[['user_id', 'high_spender']], on='user_id', how='left')
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge.columns
```

```python
# create column with median orders
ords_prods_merge['median_ordering'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#  determine frequent versus non-frequent customers
ords_prods_merge.loc[ords_prods_merge['median_ordering'] > 20, 'order_frequency'] = 'Non-frequent customer'
ords_prods_merge.loc[ords_prods_merge['median_ordering'] <= 20, 'order_frequency'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_ordering'] <= 10, 'order_frequency'] = 'Frequent customer'
```
